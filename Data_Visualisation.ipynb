```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 

# Set graphic style and figure size
sns.set(rc={'figure.figsize':(9,6)})
sns.set_style("white")
```

# DataSet

I am ussing most famous titanic dataset from keggle

```
titanic = pd.read_csv("./titanic.csv")
titanic.head()
```

# Relationships

There are two famous types of charts that are used to demonstrate relationships.

##  [Scatter plot](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)

```
sns.scatterplot(x="PassengerId", y="Age", hue="Survived", style="Sex", data=titanic)
```

## [Scatter Plot Bubble Saze](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)

```
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
sns.scatterplot(x="PassengerId", y="Age", hue="Survived", size="Age",
                      sizes=(20, 200),
                      data=titanic)
```

# Distributions 

Below are some of the most used charts to visualise distributions 

## [Histogram](https://seaborn.pydata.org/tutorial/distributions.html)

```
# titanic['Age'].apply(lambda x: int(x))
df = titanic.dropna()
sns.distplot(df['Age'], kde=True, rug=True)
```

## [Line Chart](https://seaborn.pydata.org/tutorial/distributions.html)

```
sns.distplot(df['Age'], hist=False, rug=True)
```
